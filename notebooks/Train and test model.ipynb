```python
from pathlib import Path

from src.features.as_numpy import get_data_as_numpy_array
from src.models.train import split_into_training_and_training_sets, train_model, test_model
from src.visualization.plots import get_plot_for_best_fit_line
```

```python
processed_data_file_path = Path().resolve().parents[0] / "data" / "processed" / "cleaned_housing_data.txt"
data_array = get_data_as_numpy_array(str(processed_data_file_path))
training_set, testing_set = split_into_training_and_training_sets(data_array)
slope, intercept = train_model(training_set)
print("Slope: {0}, Intercept: {1}".format(slope, intercept))
print("R Square of fit on the testing set is {0}".format(test_model(testing_set, slope, intercept)))
training_figure = get_plot_for_best_fit_line(slope, intercept, training_set[:, 0], training_set[:, 1], "Training")
training_figure.show()
testing_figure = get_plot_for_best_fit_line(slope, intercept, testing_set[:, 0], testing_set[:, 1], "Testing")
testing_figure.show()
```
